```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
students = pd.DataFrame({
    "name": ["Arnav", "kabir", "Ram", "parul", "sara"],
    "study_hours": [6, 3, 8, 2, 5],
    "attendance": [85, 70, 90, 60, 92],
    "previous_marks": [78, 55, 88, 35, 65],
    "final_marks": [82, 58, 91, 40, 70]
})

students
```

```python
feature_columns = ["study_hours", "attendance", "previous_marks"]

student_vectors = students[feature_columns].values
print(student_vectors)
```

```python

# Cosine similarity is a metric used to
#  measure the similarity between two vectors
#  by calculating the cosine of the angle between them.
#  Values closer to 1 indicate higher similarity.
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(student_vectors)

similarity_df = pd.DataFrame(
    similarity_matrix,
    index=students["name"],
    columns=students["name"]
)

similarity_df
```

```python
# Tokeniztion
# Tokenization is the process of breaking text
#  into smaller units called tokens
# text -- tokens -- token id -- embaddings -- model


```

```python
sentences = "the movie was not good"
token = sentences.lower().split()
print(token)
```

```python
vocab = {
    "the":1,
    "movie":2,
    "was":3,
    "not":4,
    "good":5

}

token_ids = [vocab[word] for word in token]
print(token_ids)
```
