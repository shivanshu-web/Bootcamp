```python
import pandas as pd
```

```python
df = pd.read_csv("session4_dummy_students_data.csv.xls")
```

```python
print(df.head())
```

```python
math_score = df["math"].to_numpy()
print(math_score)
math_avg =
print(type(math_score))
print(math_score.mean())
```

```python
df["average"] = (
    df["math"] +
    df["science"] +
    df["english"]
) / 3
```

```python
df["result"] = df["average"].apply(
    lambda x: "Pass" if x > 48 else "Fail"
)
```

```python
# Class average
class_avg = df["average"].mean()
print("Class Average:", class_avg)
```

```python
# Top student(s)
top_score = df["average"].max()
top_students = df[df["average"] == top_score]
print("\nTop Student(s):")
print(top_students[["name", "average"]])
```

```python
# Highest and lowest average
print("Highest Average:", df["average"].max())
print("Lowest Average:", df["average"].min())
```

```python
# Attendance below 75
print("\nAttendance below 75:")
print(
    df[df["attendance_percent"] < 75]
    [["name", "attendance_percent"]]
)
```

```python
# Assignments fewer than 6
print("\nAssignments below 6:")
print(
    df[df["assignments_submitted"] < 6]
    [["name", "assignments_submitted"]]
)
```

```python
# Save final report
df.to_csv("session4_student_report.csv", index=False)

print("\nFile saved successfully!")
```
